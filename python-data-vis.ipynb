## Figures, Plots, and Subplots

In Python, the best way to make a figure is by using fig objects. This can be done in the following way. Run the code below to generate a figure with an empty set of axes. You can specify the size of the figure using the `figsize` option. Type `figsize = (x, y)`, where `x` and `y` are the horizontal and vertical dimensions of the figure in inches.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (3, 2))

plt.show()
```

The reason we use the `subplots()` function here is that it makes it easy to add plots to a figure, which is commonly done in the visualization of scientific data. In the code below, we will do just that. Note that we can still use `figsize` to set the overall size of the figure, and the subplots will fit into that. Finally, you can change the internal proportions of the subplots using the `width_ratios` and `height_ratios` options.

```python
fig1, (ax1, ax2) = plt.subplots(1,2, figsize = (5,2), width_ratios = (1, 1.5))
fig2, ((ax3, ax4),(ax5, ax6)) = plt.subplots(2, 2, figsize = (6,6), height_ratios = (1.5, 1))

plt.show()
```

If you would like even more control, you can generate subplots with completely arbitrary layouts and proportions using [Gridspec](https://matplotlib.org/stable/api/_as_gen/matplotlib.gridspec.GridSpec.html).

## Scatter Plots
### Marker Styles, Colors, and Error Bars

In this section, we will cover the scatter plot, which uses the `scatter()` function in `matplotlib.pyplot`. Scatter plots are used for displaying discrete data points, where each point has a set of coordinates $(x,y)$. If you want to plot data points $(x_1, y_1), (x_2, y_2) ... (x_n, y_n)$ from lists $x = (x_1, x_2,...,x_n)$ and $y = (y_1, y_2,...,y_n)$, you can simply write 

`ax.scatter(x, y)`

to plot your data on the axis `ax`. Try this for yourself in the box below with the provided data and figure/axes.

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [1.1, 2.1, 2.8, 3.9, 5.2, 6.1, 6.8]

fig, ax = plt.subplots(figsize = (3, 3))

# Plot your data here 
ax.scatter(x, y)

plt.show()
```

This is a perfectly fine scatter plot, but we would like to customize it a bit more. To look at the options for scatter plots, you can find the full documentation for `matplotlib.pyplot.scatter()` [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html). In particular, here we will look at the ways we can customize markers. In particular, you will want to set values for 

* `s`: marker size in points x2. 
* `c`: marker color. Enter a string that could include a named color, RBG code, or hex color code. Find a full guide to specifying colors [here](https://matplotlib.org/stable/users/explain/colors/colors.html#colors-def).
* `marker`: marker style. Choose between a variety of preset options, the default being 'o' for circles. View the full list of options [here](https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers).
* `linewidths`: width of the marker outline. Enter number in pts.
* `edgecolors`: color of the marker outline. Enter as a string, similar to the value of `c`.

### Matplotlib Named Colors

See below a list of the names colors in matplotlib. Use these to customize your scatter plot!



Try adding these options to the scatter plot in the code below.

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [1.1, 2.1, 2.8, 3.9, 5.2, 6.1, 6.8]

fig, ax = plt.subplots(figsize = (3, 3))

# Plot your data here 
ax.scatter(x, y, 
           s = 40, 
           c = 'lightblue', 
           marker = '^', 
           linewidths = 1, 
           edgecolors = 'green')

plt.show()
```

### Error Bars

It is common to include error bars on a scatter plot to visualize the uncertainty of each data point. To create a scatter plot with error bars, you will need to use the `errorbar()` function instead of `scatter()`. Note that you can use `errorbar()` even if you don't want to add error bars to your plot for consistency. You can view the full documentation for `errorbar()` [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.errorbar.html). 

Use the code below to create a scatter plot with point coordinates (`x`, `y`), `x` uncertainty `dx` of 20% and `y` uncertainty `dy` with specified values. Note that you can specify a single value or an array as the uncertainty.

Customize your plot with the following parameters:

** `markersize (ms)`: marker size in points (note that it is not multiplied by two, as in `scatter()`). 
* `color (c)`: marker color. 
* `marker`: marker style. 
* `mec`: color of the marker outline. Enter as a string, similar to the value of `c`.
* `markeredgewidth (mew)` width of the marker outline in pts.
* `elinewidth`: width of the error bar line in pts.
* `ecolor`: color of the error bars.
* `linestyle (ls)`: style of the line connecting the points. To remove this line, set the value to ''.


```python
import numpy as np

x = [1, 2, 3, 4, 5, 6, 7]
y = [1.1, 2.1, 2.8, 3.9, 5.2, 6.1, 6.8]

dx = 0.2 * np.array(x) # Turn x into an array so you can multiply it by a float
dy = [0.5, 0.3, 0.5, 0.5, 0.8, 1.0, 1.3]

fig, ax = plt.subplots(figsize = (3, 3))

ax.errorbar(x, y, xerr = dx, yerr = dy,
            ms = 8,
            c = 'plum',
            marker = '^',
            mew = 1,
            mec = 'magenta',
            ecolor = 'magenta',
            elinewidth = 1,
            ls = ''
           )

plt.show()
```

## Line Plots

### Plotting Continuous Functions, Axis Scaling, Lines, and Filled Areas

Another common type of plot is a line plot. Line plots can be achieved using the `plot()` function. The full documentation for `plot()` can be found [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html). In the most basic implementation of `plot()`, the two arguments are simply the lists of arrays of x and y values.

It is important to remember that in Python, you are always plotting a discrete set of points. To create a continuous-looking line plot, simply generate a large number of points so that the line appears smooth. In the code below, play with the number of points in the `x_values` array generated by `linspace(start, stop, num_points)`.

In the code below, we will define a function $f(x)$ and plot it.

```python
def f(x):
    return np.exp(x)

n_points = 100 # Define the number of points you want to plot

x_values = np.linspace(1, 10, n_points) # Make an array of x values

y_values = f(x_values)

fig, ax = plt.subplots(figsize = (3, 3)) # Generate figure and axes

ax.plot(x_values, y_values) # Plot f(x)

plt.show()
```

As before, let's customize the line plot. The options for the line are simply the Line2D properties from matplotlib, which can be found [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D). Some key properties are 

* `color (c)`: line color, accepting the usual options for color.
* `linewidth (lw)`: line width in pts.
* `linestyle (ls)`: line style. Solid is the default, also accepts '--' for dashed, ':' for dotted, and '-.' for dash-dot (among other options).

In the code below, try to customize the color, thickness, and style of the line. 

```python
fig, ax = plt.subplots(figsize = (3, 3))

ax.plot(x_values, y_values,
       c = 'navy',
       lw = 2,
       ls = ':') 

plt.show()
```

### Axis Scaling

You may have noticed that we are plotting an exponential function. Since the y-axis of our plot spans several orders of magnitude, we may want to plot it on log-lin and a log-log axes. To do this, we simply replace `plot()` with `semilogx()`, `semilogy()` or `loglog()`. In our case, let's use `semilogy()`. By default, these functions will plot in base 10, but this can be changed using the `base` parameter. Below, try plotting the function in base 10.

```python
fig, ax = plt.subplots(figsize = (3, 3))

ax.semilogy(x_values, y_values, base = 10,
       c = 'navy',
       lw = 2,
       ls = ':') 

plt.show()
```

### Lines and Filled Areas

You may also want to add horizontal or vertical lines to your axes, or create a filled region. To do this, you can use the `hlines()`, `vlines()`, and `fill_between()` functions. The full documentation for each can be found below

* [`hlines()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hlines.html)
* [`vlines()`](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.vlines.html)
* [`fill_between()`](https://matplotlib.org/stable/gallery/lines_bars_and_markers/fill_between_demo.html)

#### Vertical and Horizontal Lines

The arguments for `hlines()` and `vlines()`  include specifying the $y$ or $x$ values where the lines are created (either as a float or array), the minim and maximum of the range over which they will be plotted, and the line colors and styles. Note that the parameters are slightly different than for `plot()`: they are called `colors` and `linestyles` since you can specify different colors and styles for each line. 

#### Filled Areas

Filled regions are useful for visualizing an uncertainty region on a continuous line plot, or for otherwise emphasizing a given region on a plot. The `fill_between(ymin, ymax)` function fills the area between the two sets of $y$ values points provided for each $x$ value. 

In the plot below, let's add an uncertainty region of 20% to our curve, as well as a vertical line at $x = 6$,



```python
fig, ax = plt.subplots(figsize = (3, 3))

ax.semilogy(x_values, y_values, base = 10,
       c = 'navy',
       lw = 2,
       ls = ':') 

ax.fill_between(x_values, 0.5 * y_values, 1.5 * y_values, 
                color = 'navy', 
                alpha = 0.2,
               linewidth = 0)

ax.vlines(6, ymin = 1, ymax = 1e4, colors = 'k')

plt.show()
```

### Colormaps

Python provides many convenient premade [colormaps](https://matplotlib.org/stable/users/explain/colors/colormaps.html). These can be used when creating plots with gradients, such as a contour plot or heatmap. They are also useful for viewing images in false color. Some color maps are specially designed for accessibility and are optimized for common forms of color blindness, as well as translating faithfully after conversion to greyscale in the case of black-and-white printing. 

Below is a usage example of a colormap in Python. The colormap is called Viridis, and is specially designed to be accessible. Click [here](https://www.youtube.com/watch?v=xAoljeRJ3lU) to watch a video of the unveiling of Viridis at SciPy 2015 if you're a big colormap nerd like me!

```python
# Heatmap 

# Generate example data (replace this with your own dataset)
data = np.random.random((5, 5))  # Replace with your data

# Create a figure and axes
fig, ax = plt.subplots()

# Create a heatmap
heatmap = ax.imshow(data, cmap='viridis')

# Add a colorbar
cbar = plt.colorbar(heatmap)
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
ax.set_title('Heatmap Example')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
np.random.seed(45)
data = [np.random.normal(0, 1, 100)]

# Custom colors for each violin

# Create a violin plot with custom colors
parts = plt.violinplot(data, positions = [1], vert = True, widths = 0.5, showmeans=False, showmedians=True)

for pc in parts['bodies']:
    pc.set_facecolor('lightblue')
    pc.set_edgecolor('royalblue')
    
# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Violin Plot with Custom Colors')

# Show the plot
plt.show()
```
