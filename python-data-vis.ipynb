## Matplotlib Documentation

Matplotlib is Python's library for visualization. It has extensive documentation available [online](https://matplotlib.org/stable/), including many [tutorials](https://matplotlib.org/stable/tutorials/index.html). Within matplotlib, you will mostly be working with pyplot to generate simple plots. You can view the documentation for pyplot [here](https://matplotlib.org/stable/api/pyplot_summary.html#module-matplotlib.pyplot). Each function within pyplot has detailed descriptions of the arguments it takes - these will be very useful when you would like to customize your plots.

## Importing matplotlib.pyplot

To import matplotlib.pyplot, simply type `import matplotlib.pyplot as plt` at the top of your code. You can then refer to the library as `plt` in your code as needed.

## General Workflow 

In general, the recipe for creating a figure is as follows:

1. Generate the data to be plotted. This is generally a list of x values and a list of y values.
2. Import matplotlib.pyplot to have access to the library and functions
3. Generate the figure and axes
4. Plot the data on the appropriate axis or axes using functions such as `plot()`, `scatter()`, `bar()`, etc..
5. Customize your plot using the built-in options for the function you used to plot.
6. Render your plot using `show()` and save or export it if you wish.

## Figures, Plots, and Subplots

In Python, the best way to make a figure is by using the [`subplots()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html) function and defining a figure and axis (or axes). The reason we use the `subplots()` function is that it makes it easy to add multiple plots/axes to a figure, which is commonly done in the visualization of scientific data. To define a figure, you can write

```
fig, ax = plt.subplots()
```

You can specify the size of the figure using the `figsize` option. Type `figsize = (x, y)`, where `x` and `y` are the horizontal and vertical dimensions of the figure in inches, inside the parentheses.

At the bottom of your code, type `plt.show()` to render your plot.

**In the code box below, generate a figure and set of axes with measurements (3, 2) and render it.**

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (3, 2))

plt.show()
````

 ### Multiple Subplots
 
As mentioned previously, `subplots()` allows you to create multiple plots within one figure. Look at the code below and notice how rows and columns of subplots are generated differently. Note that we can still use `figsize` to set the overall size of the figure, and the subplots will fit into that. Finally, you can change the internal proportions of the subplots using the `width_ratios` and `height_ratios` options. You can also force the subplots to share x or y axes using the `sharex` and `sharey` options. To create a plot with two subplots arranged horizontally, you would use the following:

```
fig1, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2)
```

Here, the `nrows` represents the number of rows, and `ncols` represents the number of columns. If you wanted to make a plot with two subplots arranged vertically, you would instead type

```
fig2, ((ax3), (ax4)) = plt.subplots(nrows = 2, ncols = 1)
```

Note how the parentheses around the axes are laid out differently when they are defined for vertical and horizontal arrangements. You can think of this as a matrix-style notation where each inner set of parentheses represents a row.

**In the box below, define a 2x2 subplot with:**
* figure size of (6, 6)
* height_ratio of (1, 1.5)
* width_ratio of (5, 3)

````python
fig3, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = (6,6), height_ratios = (1.5, 1), width_ratios = (5, 3))

plt.show()
````

If you would like even more control, you can generate subplots with completely arbitrary layouts and proportions using [Gridspec](https://matplotlib.org/stable/api/_as_gen/matplotlib.gridspec.GridSpec.html).

## Scatter Plots
### Marker Styles, Colors, Error Bars, and Title/Axis Labels

In this section, we will cover the scatter plot, which uses the `scatter()` function in `matplotlib.pyplot`. Scatter plots are used for displaying discrete data points, where each point has a set of coordinates $(x,y)$. If you want to plot data points $(x_1, y_1), (x_2, y_2) ... (x_n, y_n)$ from lists $x = (x_1, x_2,...,x_n)$ and $y = (y_1, y_2,...,y_n)$, you can simply write 

```
ax.scatter(x, y)
```
to plot your data on the axis `ax`. 

**Try this for yourself in the box below with the provided data and figure/axes.**

````python
x = [1, 2, 3, 4, 5, 6, 7]
y = [1.1, 2.1, 2.8, 3.9, 5.2, 6.1, 6.8]

fig, ax = plt.subplots(figsize = (3, 3))

# Plot your data here 
ax.scatter(x, y)

plt.show()
````

This is a perfectly fine scatter plot, but we would like to customize it a bit more. To look at the options for scatter plots, you can find the full documentation for `matplotlib.pyplot.scatter()` [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html). In particular, here we will look at the ways we can customize markers. In particular, you will want to set values for 

* `s`: marker size in points ^2 (don't ask why...). 
* `c`: marker color. Enter a string that could include a named color, RBG code, or hex color code. Find a full guide to specifying colors [here](https://matplotlib.org/stable/users/explain/colors/colors.html#colors-def).
* `marker`: marker style. Choose between a variety of preset options, the default being 'o' for circles. View the full list of options [here](https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers).
* `linewidths`: width of the marker outline. Enter number in pts.
* `edgecolors`: color of the marker outline. Enter as a string, similar to the value of `c`.

### Matplotlib Named Colors

See below a list of the names colors in matplotlib. Use these to customize your scatter plot!



**Try adding these options to the scatter plot in the code below.**

````python
x = [1, 2, 3, 4, 5, 6, 7]
y = [1.1, 2.1, 2.8, 3.9, 5.2, 6.1, 6.8]

fig, ax = plt.subplots(figsize = (3, 3))

# Plot your data here 
ax.scatter(x, y, 
           s = 49, 
           c = 'lightblue', 
           marker = '^', 
           linewidths = 1, 
           edgecolors = 'green')

plt.show()
````

### Error Bars

It is common to include error bars on a scatter plot to visualize the uncertainty of each data point. To create a scatter plot with error bars, you will need to use the `errorbar()` function instead of `scatter()`. Note that you can use `errorbar()` even if you don't want to add error bars to your plot for consistency. You can view the full documentation for `errorbar()` [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.errorbar.html). 

**Use the code below to create a scatter plot with point coordinates (`x`, `y`), `x` uncertainty `dx` of 20% and `y` uncertainty `dy` with specified values. Note that you can specify a single value or an array as the uncertainty.**

Customize your plot with the following parameters:

* `markersize (ms)`: marker size in points (note that it is not squared, as in `scatter()`). 
* `color (c)`: marker color. 
* `marker`: marker style. 
* `mec`: color of the marker outline. Enter as a string, similar to the value of `c`.
* `markeredgewidth (mew)` width of the marker outline in pts.
* `elinewidth`: width of the error bar line in pts.
* `ecolor`: color of the error bars.
* `linestyle (ls)`: style of the line connecting the points. To remove this line, set the value to ''.


````python
import numpy as np

x = [1, 2, 3, 4, 5, 6, 7]
y = [1.1, 2.1, 2.8, 3.9, 5.2, 6.1, 6.8]

dx = 0.2 * np.array(x) # Turn x into an array so you can multiply it by a float
dy = [0.5, 0.3, 0.5, 0.5, 0.8, 1.0, 1.3]

fig, ax = plt.subplots(figsize = (3, 3))

ax.errorbar(x, y, xerr = dx, yerr = dy,
            ms = 7,
            c = 'plum',
            marker = '^',
            mew = 1,
            mec = 'magenta',
            ecolor = 'magenta',
            elinewidth = 1,
            ls = ''
           )

plt.show()
````

### Titles and Axis Labels

Finally, we would like to add a title and axis labels to our plot. You can do this using the `set_title()` function, as well as the `set_xlabel()` and `set_ylabel()` functions. The argument of each function should be a string representing your desired title or label. 

**Add a title and axis labels to your plot in the code below.**

````python
import numpy as np

x = [1, 2, 3, 4, 5, 6, 7]
y = [1.1, 2.1, 2.8, 3.9, 5.2, 6.1, 6.8]

dx = 0.2 * np.array(x) # Turn x into an array so you can multiply it by a float
dy = [0.5, 0.3, 0.5, 0.5, 0.8, 1.0, 1.3]

fig, ax = plt.subplots(figsize = (3, 3))

ax.errorbar(x, y, xerr = dx, yerr = dy,
            ms = 8,
            c = 'plum',
            marker = '^',
            mew = 1,
            mec = 'magenta',
            ecolor = 'magenta',
            elinewidth = 1,
            ls = ''
           )

ax.set_title('My Title')

ax.set_xlabel('X axis label')

ax.set_ylabel('Y axis label')

plt.show()
````

## Line Plots

Another common type of plot is a line plot. Line plots can be achieved using the `plot()` function. It is important to remember that in Python, you are always plotting a discrete set of points. To create a continuous-looking line plot, simply generate a large number of points so that the line appears smooth. 

### Plotting Data with `plot()`

You can use the `plot()` function to plot real data. Let's say you have some noisy data produced by an experiment. Ours is saved in the 'data' folder in a file called 'data_with_noise.txt'. You can import the two arrays using a functuon in the numpy package.

```
x_data, y_data = np.genfromtxt('data/data_with_noise.txt', unpack = True)
```

**In the code box below, define your figure and axes, and plot your data.**


````python
fig, ax = plt.subplots(figsize = (4, 4)) # Generate figure and axes

x_data, y_data = np.genfromtxt('data/data_with_noise.txt', unpack = True)

ax.plot(x_data, y_data)

plt.show()
````

### Plotting Continuous Functions

This data looks suspiciously exponential... While fitting curves to data is beyond the scope of this lesson, let's try plotting $y = e^x$ on the same axes and see how well it matches up. To make the function look continuous, let's plot this across 100 points on the x axis. We can do this using the `linspace()` function from `numpy` to generate a list of x values.

**In the code box below, plot an exponential function overlaid with the data.**

````python
fig, ax = plt.subplots(figsize = (4, 4)) # Generate figure and axes

x_data, y_data = np.genfromtxt('data/data_with_noise.txt', unpack = True)

ax.plot(x_data, y_data)

n_points = 100 # Define the number of points you want to plot

x_values = np.linspace(1, 10, n_points) # Make an array of x values

y_values = np.exp(x_values)

ax.plot(x_values, y_values) # Plot f(x)

plt.show()
````

As before, let's customize our plot. The options for the line are simply the Line2D properties from matplotlib, which can be found [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D). Some key properties are 

* `color (c)`: line color, accepting the usual options for color.
* `linewidth (lw)`: line width in pts.
* `linestyle (ls)`: line style. Solid is the default, also accepts '--' for dashed, ':' for dotted, and '-.' for dash-dot (among other options).

**In the code below, customize the color, thickness, and style of the lines.**

````python
fig, ax = plt.subplots(figsize = (4, 4)) # Generate figure and axes

x_data, y_data = np.genfromtxt('data/data_with_noise.txt', unpack = True)

ax.plot(x_data, y_data,
        c = 'cornflowerblue',
        lw = 1)
n_points = 100 # Define the number of points you want to plot

x_values = np.linspace(1, 10, n_points) # Make an array of x values

y_values = np.exp(x_values)

ax.plot(x_values, y_values,
       c = 'navy',
       lw = 2,
       ls = '--') 

plt.show()
````

### Axis Scaling

Since the y-axis of our plot spans several orders of magnitude, we may want to plot it on a log-log or semi-log axis scale to better view our data. To do this, we simply replace `plot()` with `semilogx()`, `semilogy()` or `loglog()`. In our case, let's use `semilogy()`. By default, these functions will plot in base 10, but this can be changed using the `base` parameter. 

**Below, try plotting the function and data in base 10.**

````python
fig, ax = plt.subplots(figsize = (4, 4)) # Generate figure and axes

x_data, y_data = np.genfromtxt('data/data_with_noise.txt', unpack = True)

ax.semilogy(x_data, y_data,
        c = 'cornflowerblue',
        lw = 1)
n_points = 100 # Define the number of points you want to plot

x_values = np.linspace(1, 10, n_points) # Make an array of x values

y_values = np.exp(x_values)

ax.semilogy(x_values, y_values,
       c = 'navy',
       lw = 2,
       ls = '--') 

plt.show()
````

### Lines and Filled Areas

You may also want to add horizontal or vertical lines to your axes, or create a filled region. To do this, you can use the `hlines()`, `vlines()`, and `fill_between()` functions. The full documentation for each can be found below

* [`hlines()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hlines.html)
* [`vlines()`](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.vlines.html)
* [`fill_between()`](https://matplotlib.org/stable/gallery/lines_bars_and_markers/fill_between_demo.html)

#### Vertical and Horizontal Lines

The arguments for `hlines()` and `vlines()`  include specifying the $y$ or $x$ values where the lines are created (either as a float or array), the minim and maximum of the range over which they will be plotted, and the line colors and styles. Note that the parameters are slightly different than for `plot()`: they are called `colors` and `linestyles` since you can specify different colors and styles for each line. 

#### Filled Areas

Filled regions are useful for visualizing an uncertainty region on a continuous line plot, or for otherwise emphasizing a given region on a plot. The `fill_between(x, ymin, ymax)` function fills the area between the two sets of $y$ values points provided for each $x$ value. 

**In the plot below, let's add an uncertainty region of 50% to our curve, as well as a vertical line at $x = 8$.**

### Axis Limits

You can set the range of each axis using `set_x(y)lim(minimum, maximum)`.

**In the code below, set the x axis to be between 1 and 10, and the y axis to span the range of y values plotted.**

````python
fig, ax = plt.subplots(figsize = (4, 4)) # Generate figure and axes

x_data, y_data = np.genfromtxt('data/data_with_noise.txt', unpack = True)

ax.semilogy(x_data, y_data,
        c = 'cornflowerblue',
        lw = 1)
n_points = 100 # Define the number of points you want to plot

x_values = np.linspace(1, 10, n_points) # Make an array of x values

y_values = np.exp(x_values)

ax.semilogy(x_values, y_values,
       c = 'navy',
       lw = 2,
       ls = '--') 

ax.fill_between(x_values, 0.4 * y_values, 1.8 * y_values, 
                color = 'cornflowerblue', 
                alpha = 0.4,
               linewidth = 0)

ax.vlines(8, ymin = 1, ymax = max(y_values), colors = 'k') # the color black can be referred to as 'k'

ax.set_xlim(1, 10) # Manually set axis limits
ax.set_ylim(min(y_values), max(y_values)) # Set axis limits depending on what is being plotted

plt.show()
````

### Legends

Now that we have multiple different plots on the same axes, we may want to add a legend. To do this, we use the `legend()` function. To add a curve to the legend, you need to add a `label` when you plott it, which will automatically be reflected in the legend when it is generated. For example: `ax.plot(x, y, label = 'My Curve')` would generate a legend labeling the plotted curve as 'My Curve'. This works for any type of plotting. The full options for legends in matplotlib are shown [here](https://matplotlib.org/stable/users/explain/axes/legend_guide.html).

#### Legend Location

When you call the `legend()` function, you can specify a location for the legend. Broadly, you can assign the `loc` parameter as a string like `'upper left'` (if you do not specify, it will try to generate the legend in the best possible location). If you would like to be more specific, you can specify the legend location in the figure using the `bbox_to_anchor=(x, y)` parameter. Here, replace `x` and `y` with numerical values, where (0,0) is the lower left-hand corner and (1, 1) is the upper right-hand corner of the figure. The numbers you give it can be greater than 1 to place the legend outside the axes. 

#### Custom Legend Elements

If you would like to add a custom element to your legend, there are ways to do this using `matplotlib.patches` for color blocks, and `matplotlib.lines` for lines or markers, and then explicitly adding the patches or lines generated to the "handles" within the legend. A quick google search will reveal plenty of examples of this.

**In the code below, add labels to everything we are plotting, and then generate a legend in the upper left-hand corner of the plot.**

````python
import matplotlib.patches as mpatches

fig, ax = plt.subplots(figsize = (4, 4))

ax.semilogy(x_data, y_data, base = 10,
        c = 'cornflowerblue',
        lw = 1,
           label = 'Data')

ax.semilogy(x_values, y_values, base = 10,
       c = 'navy',
       lw = 2,
       ls = '--',
        label =  'Theory') 

ax.fill_between(x_values, 0.4 * y_values, 1.8 * y_values, 
                color = 'cornflowerblue', 
                alpha = 0.4,
                linewidth = 0,
                label = 'Uncertainty Region')

ax.vlines(8, ymin = 1, ymax = max(y_values), colors = 'k', label = 'Vertical Line') # the color black can be referred to as 'k'

ax.set_xlim(1, 10) # Manually set axis limits
ax.set_ylim(min(y_values), max(y_values)) # Set axis limits depending on what is being plotted

ax.legend(loc = 'upper left')



plt.show()
````

### Multiple Axes

Now let's say we want to add some more data to our plot by plotting the number of citations received by RU scholars in the same years. Those numbers are a much different scale than the publication data, so we would like to use a different y-axis on the right-hand side of the plot.

This can be done by generating a new axis using the `twinx()` function by typing `ax2 = ax,twinx()`. The `twiny()` function exists equivalently for creating a second horizontal axis with a shared y-axis. Then, you can plot the new data on `ax2`. Be sure to select a new color for the second dataset, as they will both get assigned to the same default color. Add a legend to differentiale between the data we are plotting.


````python
fig, ax = plt.subplots(figsize = (3, 3)) # Define figure and axes

# This is the data we would like to plot
years = [2018, 2019, 2020, 2021, 2022]
pubs = [773, 619, 630, 669, 525]
cites = [40620, 27539, 29368, 18754, 5671]

ax.bar(years, pubs, color = 'navy', alpha = 0.5)

ax2 = ax.twinx()

ax2.bar(years, cites, color = 'magenta', alpha = 0.5, label = 'Citations')

my_patch = mpatches.Patch(facecolor='navy', alpha = 0.5, edgecolor = None, label='Publications')
handles, labels = ax2.get_legend_handles_labels()
handles.append(my_patch)

ax2.legend(loc = 'upper right', handles = handles)

plt.show()
````

### Axis Labels

Our plot will be much more instructive if we add labels to the axes. To do this, we can use the `set_xlabel()` and `set_ylabel()` functions. Simply include the string you would like to turn into the axis label as the argument of the function. In the code box below, add axis labels to all 3 axes. You can use $...$ to add special characters, similar to LaTeX/Markdown.

````python
fig, ax = plt.subplots(figsize = (3, 3)) # Define figure and axes

# This is the data we would like to plot
years = [2018, 2019, 2020, 2021, 2022]
pubs = [773, 619, 630, 669, 525]
cites = [40620, 27539, 29368, 18754, 5671]

ax.bar(years, pubs, color = 'navy', alpha = 0.5)

ax2 = ax.twinx()

ax2.bar(years, cites, color = 'magenta', alpha = 0.5, label = 'Citations')

my_patch = mpatches.Patch(facecolor='navy', alpha = 0.5, edgecolor = None, label='Publications')
handles, labels = ax2.get_legend_handles_labels()
handles.append(my_patch)

ax2.legend(loc = 'upper right', handles = handles)

ax.set_ylabel('Number of Publications')
ax.set_xlabel('Year')
ax2.set_ylabel('Number of Citations ($ \mathregular{x 10}^{\mathregular{3}}$)')


plt.show()
````

### Ticks

We are now going to customize the tick locations and labels. We can do this using the `set_x(y)ticks()` and `set_x(y)ticklabels()` functions, where the inputs are list of numbers and strings, respectively. 

In the code below, set the following tick locations:
* Left vertical axis: 0, 500, 1000
* Horizontal axis: 2018, 2019, 2020, 2021, 2022
* Right vertical axis: 0, 20000, 40000

Then, set the following tick labels. We want to plot the citations in 'thousands', so set the right axis ticks to 0, 20, and 40. 
* Left vertical axis: 0, 500, 1000
* Horizontal axis: '18, '19, '20, '21, '22 (note: you must use double quotes for strings with single quotes. Ex: "'18")
* Right vertical axis: 0, 20, 40


````python
fig, ax = plt.subplots(figsize = (3, 3)) # Define figure and axes

# This is the data we would like to plot
years = [2018, 2019, 2020, 2021, 2022]
pubs = [773, 619, 630, 669, 525]
cites = [40620, 27539, 29368, 18754, 5671]

ax.bar(years, pubs, color = 'navy', alpha = 0.5)

ax2 = ax.twinx()

ax2.bar(years, cites, color = 'magenta', alpha = 0.5, label = 'Citations')

my_patch = mpatches.Patch(facecolor='navy', alpha = 0.5, edgecolor = None, label='Publications')
handles, labels = ax2.get_legend_handles_labels()
handles.append(my_patch)

ax2.legend(loc = 'upper right', handles = handles)

ax.set_ylabel('Number of Publications')
ax.set_xlabel('Year')
ax2.set_ylabel('Number of Citations ($ \mathregular{x 10}^{\mathregular{3}}$)')

# Set the tick locations
ax.set_yticks([0, 500, 1000])
ax.set_xticks([2018, 2019, 2020, 2021, 2022])
ax2.set_yticks([0, 20000, 40000])

# Adjust the tick labels
ax.set_xticklabels(["'18", "'19", "'20", "'21", "'22", ]) # Turn the list of years into strings
ax.set_yticklabels(['0', '500', '1000']) #Manually set the tick labels
ax2.set_yticklabels(['0', '20', '40'])

plt.show()
````

### Fonts

You may have noticed that all of the text in the plots we have generated so far has been the same, default Python font. If you would like to change this, you can use the `FontProperties` class to define font(s) to use in your plots. The fulll documentation for `FontProperties` can be found [here](https://matplotlib.org/stable/api/font_manager_api.html#matplotlib.font_manager.FontProperties). First, you will need to import this class using 
 
 ```
 from matplotlib.font_manager import FontProperties
```

You can then define a font in the following way. Say we want Arial, 12pt.

```
font = FontProperties(family='Arial', size = 12)

```

You can then insert this font into the various places in your code as needed by adding the following options.

#### Legend Entries

```
ax.legend(prop = font)
```

#### Title

```
ax.set_title('My Title', fontproperties = font)
```
#### Axis Labels

```
ax.set_xlabel('My X Axis Label', fontproperties = font)
ax.set_ylabel('My Y Axis Label', fontproperties = font)
```
#### Tick Labels

```
ax.set_xticklabels([list of x ticks], fontproperties = font)
ax.set_yticklabels([list of y ticks], fontproperties = font)
```
#### Legend Entries

```
ax.legend(prop = font)
```

#### A Note About Math Notation

You can use LaTeX-style math notation in python inside the strings defining axis labels and titles. To make this "math font" turn into the font you have defined, simply type `$\mathregular{ math text here}$` in the string.

In the plot below, set the font to 'Arial' and 12pt size, with 10pt in the legend.

````python
# This is what we will use to set fonts
from matplotlib.font_manager import FontProperties

font = FontProperties(family='Arial', size = 12)
font2 = FontProperties(family='Arial', size = 10)

fig, ax = plt.subplots(figsize = (3, 3)) # Define figure and axes

# This is the data we would like to plot
years = [2018, 2019, 2020, 2021, 2022]
pubs = [773, 619, 630, 669, 525]
cites = [40620, 27539, 29368, 18754, 5671]

ax.bar(years, pubs, color = 'navy', alpha = 0.5)

ax2 = ax.twinx()

ax2.bar(years, cites, color = 'magenta', alpha = 0.5, label = 'Citations')

my_patch = mpatches.Patch(facecolor='navy', alpha = 0.5, edgecolor = None, label='Publications')
handles, labels = ax2.get_legend_handles_labels()
handles.append(my_patch)

ax2.legend(loc = 'upper right', handles = handles, prop = font2)

ax.set_ylabel('Number of Publications', fontproperties = font)
ax.set_xlabel('Year', fontproperties = font)
ax2.set_ylabel('Number of Citations ($ \mathregular{x 10}^{\mathregular{3}}$)', fontproperties = font)

# Set the tick locations
ax.set_yticks([0, 500, 1000])
ax.set_xticks([2018, 2019, 2020, 2021, 2022])
ax2.set_yticks([0, 20000, 40000])

# Adjust the tick labels
ax.set_xticklabels(["'18", "'19", "'20", "'21", "'22", ], fontproperties = font) # Turn the list of years into strings
ax.set_yticklabels(['0', '500', '1000'], fontproperties = font) #Manually set the tick labels
ax2.set_yticklabels(['0', '20', '40'], fontproperties = font)

plt.show()
````
