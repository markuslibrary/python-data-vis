In this workshop you will be introduced to some of Python's utilities for data visualization. By then end, you should know how to display your data in publication-worthy figures, and how to navigate the matplotlib documentation to fully customize your plots.

## Matplotlib Documentation

Matplotlib is Python's library for visualization. It has extensive documentation available [online](https://matplotlib.org/stable/), including many [tutorials](https://matplotlib.org/stable/tutorials/index.html). Within matplotlib, you will mostly be working with pyplot to generate simple plots. ## Importing matplotlib.pyplot

To import matplotlib.pyplot, simply type `import matplotlib.pyplot as plt` at the top of your code. You can then refer to the library as `plt` in your code as needed.

## General Workflow for Creating Figures

In general, the recipe for creating a figure is as follows:

1. Import matplotlib.pyplot to have access to the library and functions
2. Generate the figure and axes
3. Generate the data to be plotted. This is generally a list of x values and a list of y values.
4. Plot the data on the appropriate axis or axes using functions such as `plot()`, `scatter()`, `bar()`, etc..
5. Customize your plot using the built-in options for the function you used to plot.
6. Render your plot using `show()` and save or export it if you wish.

## Figures, Plots, and Subplots

In Python, the best way to make a figure is by using the [`subplots()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html) function to define a figure and set(s) of axes. The reason we use the `subplots()` function is that it makes it easy to add multiple plots/axes to a figure, which is commonly done in the visualization of scientific data. To define a figure, you can write

```
fig, ax = plt.subplots()
```

Here, `fig` is your figure, and `ax` is your set of axes where you will plot your data. You can specify the size of the figure using the `figsize` option. Type `figsize = (x, y)`, where `x` and `y` are the horizontal and vertical dimensions of the figure in inches, inside the parentheses.

At the bottom of your code, type `plt.show()` to render your plot.

**In the code box below, generate a figure and set of axes with measurements (3, 2) and render it.**

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (3, 2))

plt.show()
````

 ### Multiple Subplots
 
As mentioned previously, `subplots()` allows you to create multiple plots within one figure. Look at the code below and notice how rows and columns of subplots are generated differently. Note that we can still use `figsize` to set the overall size of the figure, and the subplots will fit into that. Finally, you can change the internal proportions of the subplots using the `width_ratios` and `height_ratios` options. You can also force the subplots to share x or y axes using the `sharex` and `sharey` options. To create a plot with two subplots arranged horizontally, you would use the following:

```
fig1, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2)
```

Here, the `nrows` represents the number of rows, and `ncols` represents the number of columns. If you wanted to make a plot with two subplots arranged vertically, you would instead type

```
fig2, ((ax3), (ax4)) = plt.subplots(nrows = 2, ncols = 1)
```

Note how the parentheses around the axes are laid out differently when they are defined for vertical and horizontal arrangements. You can think of this as a matrix-style notation where each inner set of parentheses represents a row.

**In the box below, define a 2x2 subplot with `figsize`, `height_ratios`, and `width_ratios` specified.**

````python
fig3, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = (6,6), height_ratios = (1.5, 1), width_ratios = (5, 3))

plt.show()
````

If you would like even more control, you can generate subplots with completely arbitrary layouts and proportions using [Gridspec](https://matplotlib.org/stable/api/_as_gen/matplotlib.gridspec.GridSpec.html).

## Scatter Plots

Scatter plots are used for displaying discrete data points, where each point has a set of coordinates $(x,y)$. If you want to plot data points $(x_1, y_1), (x_2, y_2) ... (x_n, y_n)$ from lists $x = (x_1, x_2,...,x_n)$ and $y = (y_1, y_2,...,y_n)$, you can simply write 

```
ax.scatter(x, y)

to plot your data on the axis `ax`. You can use the following lists to create your scatter plot:

```
x = [1, 2, 3, 4, 5, 6, 7]
y = [1.1, 2.1, 2.8, 3.9, 5.2, 6.1, 6.8]

**Try this for yourself in the box below with the provided data and figure/axes.**

First, define your figure and axes, and then use `scatter()` to plot the data. Remember to type `show()` at the end!

````python
# Define your data
x = [1, 2, 3, 4, 5, 6, 7]
y = [1.1, 2.1, 2.8, 3.9, 5.2, 6.1, 6.8]

# Define your figure
fig, ax = plt.subplots(figsize = (3, 3))

# Plot your data here 
ax.scatter(x, y)

plt.show()
````

This is a perfectly fine scatter plot, but we would like to customize it a bit more. To look at the options for scatter plots, you can find the full documentation for `matplotlib.pyplot.scatter()` [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html). * `s`: marker size in points ^2 (don't ask why...). 
* `c`: marker color. Enter a string that could include a named color, RBG code, or hex color code. Find a full guide to specifying colors [here](https://matplotlib.org/stable/users/explain/colors/colors.html#colors-def).
* `marker`: marker style. Choose between a variety of preset options, the default being 'o' for circles. View the full list of options [here](https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers).
* `linewidths`: width of the marker outline. Enter number in pts.
* `edgecolors`: color of the marker outline. Enter as a string, similar to the value of `c`.

### Matplotlib Named Colors

See below a list of the names colors in matplotlib. Use these to customize your scatter plot!

### Error Bars

It is common to include error bars on a scatter plot to visualize the uncertainty of each data point. To create a scatter plot with error bars, you will need to use the `errorbar()` function instead of `scatter()`. Note that you can use `errorbar()` even if you don't want to add error bars to your plot for consistency. You can view the full documentation for `errorbar()` [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.errorbar.html). 

In addition to the usual `x` and `y` arguments, `errorbar()` takes **optional** `xerr` and `yerr` arguments, representing the horizontal and vertical error bars. These can be just a number, or a list matching the length of the data if you want individually specified error bars.

You can use:

```
dx = 0.2 * np.array(x) # Turn x into an array so you can multiply it by a float
dy = [0.5, 0.3, 0.5, 0.5, 0.8, 1.0, 1.3]
```

**Remember to import numpy as np!!**

Customize your plot with the following parameters:

* `markersize (ms)`: marker size in points (note that it is not squared, as in `scatter()`). 
* `markerfacecolor (mfc)`: marker face color. 
* `color (c)`: Color of the line connecting your points, if any. If there is no line, you can use `color` to specify the face color of the markers.
* `marker`: marker style. 
* `markeredgecolor (mec)`: color of the marker outline. Enter as a string, similar to the value of `c`.
* `markeredgewidth (mew)` width of the marker outline in pts.
* `elinewidth`: width of the error bar line in pts.
* `ecolor`: color of the error bars.
* `linestyle (ls)`: style of the line connecting the points. To remove this line, set the value to ''.

**Use the code below to create a scatter plot with point coordinates (`x`, `y`), `x` uncertainty `dx` of 20% and `y` uncertainty `dy` with specified values. Customize your markers.**



````python
import numpy as np

# Define your data
x = [1, 2, 3, 4, 5, 6, 7]
y = [1.1, 2.1, 2.8, 3.9, 5.2, 6.1, 6.8]

# Define your figure
fig, ax = plt.subplots(figsize = (3, 3))

dx = 0.2 * np.array(x) # Turn x into an array so you can multiply it by a float
dy = [0.5, 0.3, 0.5, 0.5, 0.8, 1.0, 1.3]

ax.errorbar(x, y, xerr = dx, yerr = dy,
            ms = 7,
            mfc = 'plum',
            marker = '^',
            mew = 1,
            mec = 'magenta',
            ecolor = 'magenta',
            elinewidth = 1,
            ls = ''
           )

plt.show()
````

### Titles and Axis Labels

Finally, we would like to add a title and axis labels to our plot. You can do this using the `set_title()` function, as well as the `set_xlabel()` and `set_ylabel()` functions, acting on your axis. The argument of each function should be a string representing your desired title or label. 

**Add a title and axis labels to your plot in the code below.**

````python
import numpy as np

# Define your data
x = [1, 2, 3, 4, 5, 6, 7]
y = [1.1, 2.1, 2.8, 3.9, 5.2, 6.1, 6.8]

# Define your figure
fig, ax = plt.subplots(figsize = (3, 3))

dx = 0.2 * np.array(x) # Turn x into an array so you can multiply it by a float
dy = [0.5, 0.3, 0.5, 0.5, 0.8, 1.0, 1.3]

ax.errorbar(x, y, xerr = dx, yerr = dy,
            ms = 7,
            mfc = 'plum',
            marker = '^',
            mew = 1,
            mec = 'magenta',
            ecolor = 'magenta',
            elinewidth = 1,
            ls = ''
           )

ax.set_title('My Title')

ax.set_xlabel('X axis label')

ax.set_ylabel('Y axis label')

plt.show()
````

### Plotting Continuous Functions

This data looks suspiciously exponential... While fitting curves to data is beyond the scope of this lesson, let's try plotting $y = e^x$ on the same axes and see how well it matches up. It is important to remember that in Python, you are always plotting a discrete set of points even if you are trying to visualize a continuous function. To create a continuous-looking line plot, simply generate a large number of points so that the line appears smooth. To make the function look continuous, let's plot this across 100 points on the x axis. We can do this using the `linspace()` function from `numpy` to generate a list of x values.

**In the code box below, plot an exponential function overlaid with the data.**

````python
# Generate figure and axes
fig, ax = plt.subplots(figsize = (4, 4)) # Generate figure and axes

# Import data
x_data, y_data = np.genfromtxt('data/data_with_noise.txt', unpack = True)

# Plot data
ax.plot(x_data, y_data)

# Define the number of points you want to plot
n_points = 100

# Generate x values
x_values = np.linspace(1, 10, n_points) 

# Generate y values
y_values = np.exp(x_values)

# Plot function
ax.plot(x_values, y_values) 

plt.show()
````

As before, let's customize our plot. The options for the line are simply the Line2D properties from matplotlib, which can be found [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D). Some key properties are 

* `color (c)`: line color, accepting the usual options for color.
* `linewidth (lw)`: line width in pts.
* `linestyle (ls)`: line style. Solid is the default, also accepts '--' for dashed, ':' for dotted, and '-.' for dash-dot (among other options).

**In the code below, customize the color, thickness, and style of the lines.**

````python
# Generate figure and axes
fig, ax = plt.subplots(figsize = (4, 4)) # Generate figure and axes

# Import data
x_data, y_data = np.genfromtxt('data/data_with_noise.txt', unpack = True)

# Plot data
ax.plot(x_data, y_data, c = 'cornflowerblue', lw = 1)

# Define the number of points you want to plot
n_points = 100

# Generate x values
x_values = np.linspace(1, 10, n_points) 

# Generate y values
y_values = np.exp(x_values)

# Plot function
ax.plot(x_values, y_values, c = 'navy', lw = 2, ls = '--') 

plt.show()

````

### Lines and Filled Areas

You may also want to add horizontal or vertical lines to your axes, or create a filled region. To do this, you can use the [`hlines()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hlines.html),  [`vlines()`](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.vlines.html), and [`fill_between()`](https://matplotlib.org/stable/gallery/lines_bars_and_markers/fill_between_demo.html) functions. 

#### Vertical and Horizontal Lines

The arguments for `hlines()` and `vlines()`  include specifying the $y$ or $x$ values where the lines are created (either as a float or array), the minim and maximum of the range over which they will be plotted, and the line colors and styles. Note that the parameters are slightly different than for `plot()`: they are called `colors` and `linestyles` since you can specify different colors and styles for each line. 

#### Filled Areas

Filled regions are useful for visualizing an uncertainty region on a continuous line plot, or for otherwise emphasizing a given region on a plot. The `fill_between(x, ymin, ymax)` function fills the area between the two sets of $y$ values points provided for each $x$ value. 

**In the plot below, let's add an uncertainty region of 0.4 to 1.8 of our curve, as well as a vertical line at $x = 8$.**

### Axis Limits

You can set the range of each axis using `set_x(y)lim(minimum, maximum)`.

**In the code below, set the x axis to be between 1 and 10, and the y axis to span the range of y values plotted.**

````python
# Generate figure and axes
fig, ax = plt.subplots(figsize = (4, 4)) # Generate figure and axes

# Import data
x_data, y_data = np.genfromtxt('data/data_with_noise.txt', unpack = True)

# Plot data
ax.semilogy(x_data, y_data, c = 'cornflowerblue', lw = 1)

# Define the number of points you want to plot
n_points = 100

# Generate x values
x_values = np.linspace(1, 10, n_points) 

# Generate y values
y_values = np.exp(x_values)

# Plot function
ax.semilogy(x_values, y_values, c = 'navy', lw = 2, ls = '--') 

# Filled area
ax.fill_between(x_values, 0.4 * y_values, 1.8 * y_values, 
                color = 'cornflowerblue', 
                alpha = 0.4,
               linewidth = 0)

# Vertical line
ax.vlines(8, ymin = 1, ymax = max(y_data), colors = 'k') # the color black can be referred to as 'k'

# x and y limits
ax.set_xlim(1, 10) 
ax.set_ylim(min(y_data), max(y_data)) 

plt.show()
````

### Legends

Now that we have multiple different plots on the same axes, we may want to add a legend. To do this, we use the `legend()` function. To add a curve to the legend, you need to add a `label` when you plot it, which will automatically be reflected in the legend when it is generated. For example: `ax.plot(x, y, label = 'My Curve')` would generate a legend labeling the plotted curve as 'My Curve'. This works for any type of plotting. The full options for legends in matplotlib are shown [here](https://matplotlib.org/stable/users/explain/axes/legend_guide.html).

#### Legend Location

When you call the `legend()` function, you can specify a location for the legend. Broadly, you can assign the `loc` parameter as a string like `'upper left'` (if you do not specify, it will try to generate the legend in the best possible location). If you would like to be more specific, you can specify the legend location in the figure using the `bbox_to_anchor=(x, y)` parameter. Here, replace `x` and `y` with numerical values, where (0,0) is the lower left-hand corner and (1, 1) is the upper right-hand corner of the figure. The numbers you give it can be greater than 1 to place the legend outside the axes. 

#### Custom Legend Elements

If you would like to add a custom element to your legend, there are ways to do this using `matplotlib.patches` for color blocks, and `matplotlib.lines` for lines or markers, and then explicitly adding the patches or lines generated to the "handles" within the legend. A quick google search will reveal plenty of examples of this.

**In the code below, add labels to everything we are plotting, and then generate a legend in the upper left-hand corner of the plot.**

````python
# Generate figure and axes
fig, ax = plt.subplots(figsize = (4, 4)) # Generate figure and axes

# Import data
x_data, y_data = np.genfromtxt('data/data_with_noise.txt', unpack = True)

# Plot data
ax.semilogy(x_data, y_data, c = 'cornflowerblue', lw = 1, label = 'Data')

# Define the number of points you want to plot
n_points = 100

# Generate x values
x_values = np.linspace(1, 10, n_points) 

# Generate y values
y_values = np.exp(x_values)

# Plot function
ax.semilogy(x_values, y_values, c = 'navy', lw = 2, ls = '--', label = 'Theory') 

# Filled area
ax.fill_between(x_values, 0.4 * y_values, 1.8 * y_values, 
                color = 'cornflowerblue', 
                alpha = 0.4,
               linewidth = 0,
               label = 'Uncertainty Region')

# Vertical line
ax.vlines(8, ymin = 1, ymax = max(y_data), colors = 'k', label = 'Line') # the color black can be referred to as 'k'

# x and y limits
ax.set_xlim(1, 10) 
ax.set_ylim(min(y_data), max(y_data)) 

ax.legend(loc = 'upper left')

plt.show()
````

### Fonts

You may have noticed that all of the text in the plots we have generated so far has been the same, default Python font. If you would like to change this, you can use the `FontProperties` class to define font(s) to use in your plots. The fulll documentation for `FontProperties` can be found [here](https://matplotlib.org/stable/api/font_manager_api.html#matplotlib.font_manager.FontProperties). First, you will need to import this class using 
 
 ```
 from matplotlib.font_manager import FontProperties
```

You can then define a font in the following way. Say we want Arial, 12pt.

```
font = FontProperties(family='Arial', size = 12)

```

You can then insert this font into the various places in your code as needed by adding the following options.

#### Legend Entries

```
ax.legend(prop = font)
```

#### Title

```
ax.set_title('My Title', fontproperties = font)
```
#### Axis Labels

```
ax.set_xlabel('My X Axis Label', fontproperties = font)
ax.set_ylabel('My Y Axis Label', fontproperties = font)
```
#### Tick Labels

```
ax.set_xticklabels([list of x ticks], fontproperties = font)
ax.set_yticklabels([list of y ticks], fontproperties = font)
```
#### Legend Entries

```
ax.legend(prop = font)
```

**In the plot below, set the font to 'Arial' and 12pt size.**

Define a font and apply it to all of the text in the plot. For the tick labels, you will need to use:

```
for label in ax.get_yticklabels():
    label.set_fontproperties(font)
```

This allows you to specify the font for the tick labels without actually specifying the labels themselves. Of course, you could also choose the tick locations and labels explicitly and then specify the font there.

````python
from matplotlib.font_manager import FontProperties

# Define fonts
font = FontProperties(family='Arial', size = 12)

# Import data
month, rain, snow, t_avg, t_max, t_min = np.genfromtxt('data/weather_data.csv', unpack = True, delimiter = ',', skip_header = 1)

# Generate figure and axes
fig, ((ax1), (ax2)) = plt.subplots (ncols = 1, nrows = 2, sharex = True, figsize = (4.5, 4.5))

# Plot Data
ax1.errorbar(month, t_max, marker = 'o', linewidth = 1, markerfacecolor = 'mistyrose', markeredgecolor = 'coral', color = 'coral', label = 'Max.')
ax1.errorbar(month, t_avg, marker = 'o', linewidth = 1, markerfacecolor = 'white', markeredgecolor = 'grey', color = 'grey', label = 'Avg.')
ax1.errorbar(month, t_min, marker = 'o', linewidth = 1, markerfacecolor = 'lightsteelblue', markeredgecolor = 'royalblue', color = 'royalblue', label = 'Min.')

ax2.bar(month, rain, color = 'lightsteelblue', align = 'edge', width = 0.3, label = 'Rain', edgecolor = 'royalblue')
ax2.bar(month, snow, color = 'ghostwhite', align = 'edge', width = -0.3, label = 'Snow', edgecolor = 'lightsteelblue')

#Create filled regions
ax1.fill_between(month, t_min, t_avg, color = 'lightsteelblue', alpha = 0.3)
ax1.fill_between(month, t_avg, t_max, color = 'mistyrose', alpha = 0.5)

# Create legends
ax1.legend(prop = font, loc = 'upper left')
ax2.legend(prop = font, loc = 'upper right')

# Set ax1 y limits so legend fits nicely
ax1.set_ylim(20, 110)

# Set y axis labels and title
ax1.set_ylabel('Temperature (F)', fontproperties = font)
ax2.set_ylabel('Precipitation (in)', fontproperties = font)
ax1.set_title('NYC Weather 1981 - 2010', fontproperties = font)

# Set month ticks and labels
ax1.set_xticks(month)
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
ax2.set_xticklabels(month_names, fontproperties = font, rotation = 45)

# Make ticks on all sides, pointing inwards, make slightly shorter than default
ax1.tick_params(direction = 'in', axis = 'both', top = True, right = True, length = 3)
ax2.tick_params(direction = 'in', axis = 'both', top = True, right = True, length = 3)

# Set tick label fonts
for label in ax1.get_yticklabels():
    label.set_fontproperties(font)

for label in ax2.get_yticklabels():
    label.set_fontproperties(font)
    
````
