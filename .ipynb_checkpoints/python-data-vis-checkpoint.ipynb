## Histograms

Use data from the NYC 2015 Street Tree Census

```python
import numpy as np

tree_dbh = np.genfromtxt('data/2015_Street_Tree_Census_Tree_Data.csv', unpack = True, usecols = 3, delimiter = ',', skip_header = 1)

# Define figure and axes
fig, ax = plt.subplots()

# Create a histogram
ax.hist(tree_dbh, bins=10, color='skyblue', edgecolor='black')

ax.set_xlim(0, 60)
```

## Scatter plots with and without error bars

Take a second to look at the [scatter plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html) and [error bar](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.errorbar.html) documentation. Note the various parameters that you can apply to each. We will now plot the data listed below using both the `scatter()` and `errorbar()` functions in matplotlib.




```python
import numpy as np
import matplotlib.pyplot as plt

# Define data
x = [1.5, 3.3, 2.1, 4.9, 5.7, 8.2, 9.4]
y = [0.2, 4.3, 6.8, 4.5, 6.1, 7.2, 8.6]

# Uncertainties for x and y data
dx = 0.5
dy = [0.4, 0.5, 0.5, 0.6, 0.4, 0.5, 0.4]

# Define figure and axes
fig, (ax1, ax2) = plt.subplots(2)

ax1.scatter(x, y)

ax2.errorbar(x, y, xerr = dx, yerr = dy, ls = 'none', marker = 'o', linewidth  = 1, color = 'orange', mec = 'red', mfc = 'pink')



```
